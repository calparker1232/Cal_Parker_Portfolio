```python
import pandas as pd
Options_Data_CSV=pd.read_csv('ET Option Data-8-24-10-04-24.csv')
Options_Data_CSV_df=pd.DataFrame(Options_Data_CSV)

XPrice = 16
First_Month = '12/20/2024'
Second_Month = '01/17/2025'
#puts
puts_group = Options_Data_CSV_df.groupby('data.type').get_group('put')
#Strike Price
puts_group_XPrice = puts_group.groupby('data.strike').get_group(XPrice)

# Convert 'data.expiration' to datetime objects
puts_group_XPrice['data.expiration'] = pd.to_datetime(puts_group_XPrice['data.expiration'])

# Filter for the two expiration dates using the correct datetime format
puts_group_XPrice_first_expiration = puts_group_XPrice.groupby('data.expiration').get_group(pd.to_datetime(First_Month))
puts_group_XPrice_second_expiration = puts_group_XPrice.groupby('data.expiration').get_group(pd.to_datetime(Second_Month))
```

```python
puts_group_XPrice_second_expiration_modified_table = puts_group_XPrice_second_expiration[['data.ask','data.bid','data.mark','data.expiration','data.date']]
puts_group_XPrice_first_expiration_modified_table = puts_group_XPrice_first_expiration[['data.ask','data.bid','data.mark','data.expiration','data.date']]
```

```python
# Calculate the difference between 'data.ask' and 'data.mark' for the second expiration table
puts_group_XPrice_second_expiration_modified_table['ask_mark_difference'] = (
    puts_group_XPrice_second_expiration_modified_table['data.ask'] - puts_group_XPrice_second_expiration_modified_table['data.mark']
)

# Calculate the difference between 'data.ask' and 'data.mark' for the first expiration table
puts_group_XPrice_first_expiration_modified_table['ask_mark_difference'] = (
    puts_group_XPrice_first_expiration_modified_table['data.ask'] - puts_group_XPrice_first_expiration_modified_table['data.mark']
)

# Calculate the difference between 'data.ask' and 'data.mark' for the second expiration table
puts_group_XPrice_second_expiration_modified_table['bid_mark_difference'] = (
    puts_group_XPrice_second_expiration_modified_table['data.mark'] - puts_group_XPrice_second_expiration_modified_table['data.bid']
)

# Calculate the difference between 'data.ask' and 'data.mark' for the first expiration table
puts_group_XPrice_first_expiration_modified_table['bid_mark_difference'] = (
    puts_group_XPrice_first_expiration_modified_table['data.mark'] - puts_group_XPrice_first_expiration_modified_table['data.bid']
)

```

```python
import matplotlib.pyplot as plt

# Function to plot ask, mark, and ask-mark difference over time
def plot_ask_chart(df, title):
    plt.figure(figsize=(10, 6))
    plt.plot(df['data.date'], df['data.ask'], label='Ask Price', marker='o')
    plt.plot(df['data.date'], df['data.mark'], label='Mark Price', marker='x')
    plt.plot(df['data.date'], df['ask_mark_difference'], label='Ask-Mark Difference', linestyle='--')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(f'Ask, Mark, and Ask-Mark Difference over Time - {title}')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Now plot the charts for each DataFrame

# Chart for the second expiration (ask)
plot_ask_chart(second_expiration_result_askDiff, "Second Expiration (Ask)")

# Chart for the first expiration (ask)
plot_ask_chart(first_expiration_result_askDiff, "First Expiration (Ask)")
```

```python
import matplotlib.pyplot as plt

# Function to plot bid, mark, and bid-mark difference over time
def plot_bid_chart(df, title):
    plt.figure(figsize=(10, 6))
    plt.plot(df['data.date'], df['data.bid'], label='Bid Price', marker='o')
    plt.plot(df['data.date'], df['data.mark'], label='Mark Price', marker='x')
    plt.plot(df['data.date'], df['bid_mark_difference'], label='Bid-Mark Difference', linestyle='--')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(f'Bid, Mark, and Bid-Mark Difference over Time - {title}')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Now plot the charts for each DataFrame

# Chart for the second expiration (bid)
plot_bid_chart(second_expiration_result_bidDiff, "Second Expiration (Bid)")

# Chart for the first expiration (bid)
plot_bid_chart(first_expiration_result_bidDiff, "First Expiration (Bid)")
```

```python
# Merging first expiration askDiff and bidDiff on 'data.date'
merged_first_expiration = pd.merge(
    first_expiration_result_askDiff,
    first_expiration_result_bidDiff,
    on='data.date',
    suffixes=('_ask', '_bid')
)

# Merging second expiration askDiff and bidDiff on 'data.date'
merged_second_expiration = pd.merge(
    second_expiration_result_askDiff,
    second_expiration_result_bidDiff,
    on='data.date',
    suffixes=('_ask', '_bid')
)

# Now merging the first expiration and second expiration data frames
# This will merge on 'data.date' and avoid duplicate 'data.mark' columns
final_merged_result_df = pd.merge(
    merged_first_expiration,
    merged_second_expiration,
    on='data.date',
    suffixes=('_first_exp', '_second_exp')
)

# Display the final merged DataFrame
final_merged_result_df

```
