```python
import pandas as pd
from statsmodels.tsa.stattools import coint
import matplotlib.pyplot as plt


NIO_Options_Data_CSV=pd.read_csv('NIO Options Data2.csv')
NIO_Options_Data_CSV_df=pd.DataFrame(NIO_Options_Data_CSV)
NIO_Options_Data_CSV_df.head(5)
# Load the NIO options data from the CSV file
#NIO_Options_Data_CSV = pd.read_csv("NIO Options Data.csv")

# Filter data for puts with strike price of 6.00 and expiration dates in November and December 2024
puts_november = NIO_Options_Data_CSV_df[
    (NIO_Options_Data_CSV_df['data.type'] == 'put') &
    (NIO_Options_Data_CSV_df['data.strike'] == 6.0) &
    (NIO_Options_Data_CSV_df['data.expiration'] == '11/15/2024')
]

puts_december = NIO_Options_Data_CSV_df[
    (NIO_Options_Data_CSV_df['data.type'] == 'put') &
    (NIO_Options_Data_CSV_df['data.strike'] == 6.00) &
    (NIO_Options_Data_CSV_df['data.expiration'] == '12/20/2024')
]

# Ensure both datasets are sorted by date and have the same date range
puts_november_df = puts_november[['data.date', 'data.bid']]
puts_december_df = puts_december[['data.date', 'data.ask']]


print(puts_november_df.head())
print(puts_december_df.head())
```

```python
import pandas as pd

# Merge the November and December data on 'data.date'
merged_df = pd.merge(puts_november_df, puts_december_df, on='data.date', how='inner')

# Calculate the difference between December ask and November bid
merged_df['difference'] = merged_df['data.ask'] - merged_df['data.bid']

# Display the result
print(merged_df[['data.date', 'data.bid', 'data.ask', 'difference']])

# Plotting the data
plt.figure(figsize=(10, 6))

# Plot November Bid
plt.plot(merged_df['data.date'], merged_df['data.bid'], label='November Bid', marker='o')

# Plot December Ask
plt.plot(merged_df['data.date'], merged_df['data.ask'], label='December Ask', marker='o')

# Plot the difference
plt.plot(merged_df['data.date'], merged_df['difference'], label='Difference (Dec Ask - Nov Bid)', marker='o')

# Add titles and labels
plt.title('November Bid vs December Ask Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose

# Ensure 'data.date' is in datetime format
merged_df['data.date'] = pd.to_datetime(merged_df['data.date'])

# Set 'data.date' as the index for the time series
merged_df.set_index('data.date', inplace=True)

# Perform seasonal decomposition on the 'difference' variable
decomposition = seasonal_decompose(merged_df['difference'], model='additive', period=5)  # Adjust 'period' based on your data

# Extract the trend, seasonal, and residual components
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Plot the decomposition results
plt.figure(figsize=(10, 8))

# Plot original data
plt.subplot(411)
plt.plot(merged_df['difference'], label='Original Difference', marker='o')
plt.legend(loc='best')

# Plot trend
plt.subplot(412)
plt.plot(trend, label='Trend', marker='o')
plt.legend(loc='best')

# Plot seasonal component
plt.subplot(413)
plt.plot(seasonal, label='Seasonal', marker='o')
plt.legend(loc='best')

# Plot residuals
plt.subplot(414)
plt.plot(residual, label='Residual', marker='o')
plt.legend(loc='best')

plt.tight_layout()
plt.show()

# Print the decomposed components
print("Trend Component:")
print(trend.dropna().head())
print("\nSeasonal Component:")
print(seasonal.dropna().head())
print("\nResidual Component:")
print(residual.dropna().head())
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a histogram of the 'difference' variable
plt.figure(figsize=(10, 6))
plt.hist(merged_df['difference'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Difference (Dec Ask - Nov Bid)')
plt.xlabel('Difference')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Calculate the mean and standard deviation of the 'difference' variable
mean_diff = merged_df['difference'].mean()
std_diff = merged_df['difference'].std()

# Plot the 'difference' values over time
plt.figure(figsize=(10, 6))
plt.plot(merged_df.index, merged_df['difference'], label='Difference', marker='o', color='blue')

# Add mean and standard deviation lines
plt.axhline(mean_diff, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_diff:.2f}')
plt.axhline(mean_diff + std_diff, color='green', linestyle='--', linewidth=1, label=f'Mean + 1 STD: {mean_diff + std_diff:.2f}')
plt.axhline(mean_diff - std_diff, color='green', linestyle='--', linewidth=1, label=f'Mean - 1 STD: {mean_diff - std_diff:.2f}')
plt.axhline(mean_diff + 2 * std_diff, color='orange', linestyle='--', linewidth=1, label=f'Mean + 2 STD: {mean_diff + 2 * std_diff:.2f}')
plt.axhline(mean_diff - 2 * std_diff, color='orange', linestyle='--', linewidth=1, label=f'Mean - 2 STD: {mean_diff - 2 * std_diff:.2f}')

# Customize the plot
plt.title('Difference (Dec Ask - Nov Bid) with Standard Deviation Markers')
plt.xlabel('Date')
plt.ylabel('Difference')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()

# Print the mean and standard deviation for reference
print(f"Mean of Difference: {mean_diff:.2f}")
print(f"Standard Deviation of Difference: {std_diff:.2f}")
```
